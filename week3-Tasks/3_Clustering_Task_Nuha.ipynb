## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
from scipy import stats

def remove_outlier_zscore(df, threshold=3):
  numeric_df = df.select_dtypes(['number'])
  z_scores = stats.zscore(numeric_df)
  abs_z_scores = abs(z_scores)

  outliers = (abs_z_scores > threshold).any(axis=1)

  df_cleaned = df[~outliers]
  return df_cleaned

df = remove_outlier_zscore(df, threshold=3)
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.title('Box Plot after removing outlier')
plt.show()
```

```python
df["Accident_Severity"].fillna(df['Accident_Severity'].mode()[0], inplace=True)
df["Number_of_Casualties"].fillna(df['Number_of_Casualties'].mode()[0], inplace=True)
df["Light_Conditions"].fillna(df['Light_Conditions'].mode()[0], inplace=True)
df["Urban_or_Rural_Area"].fillna(df['Urban_or_Rural_Area'].mode()[0], inplace=True)
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
from sklearn.cluster import KMeans

k_values = range(1, 11)
wcss = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=9, random_state=0)
kmeans.fit(x_pca)

cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=labels, marker='o')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red')
plt.title('K-Means Clustering')

plt.grid(True)
plt.legend()
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
from sklearn.metrics import silhouette_score
labels = kmeans.labels_
silhouette_avg = silhouette_score(x_pca, labels)
print(f'Silhouette Score: {silhouette_avg:.3f}')
```
